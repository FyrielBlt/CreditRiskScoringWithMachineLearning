```python
def tableau_valeurs_manquantes(df):
    # Nombre total de valeurs manquantes
    val_manquantes = df.isnull().sum()
    
    # Pourcentage de valeurs manquantes
    pourcentage_val_manquantes = 100 * df.isnull().sum() / len(df)
    
    # Création d'un tableau avec les résultats
    tableau_val_manquantes = pd.concat([val_manquantes, pourcentage_val_manquantes], axis=1)
    
    # Renommer les colonnes
    tableau_val_manquantes_ren_colonnes = tableau_val_manquantes.rename(
        columns={0: 'Valeurs Manquantes', 1: '% du Total des Valeurs'})
    
    # Trier le tableau par pourcentage de valeurs manquantes de façon descendante
    tableau_val_manquantes_ren_colonnes = tableau_val_manquantes_ren_colonnes[
        tableau_val_manquantes_ren_colonnes.iloc[:, 1] != 0].sort_values(
            '% du Total des Valeurs', ascending=False).round(1)
    
    # Afficher quelques informations de synthèse
    print("Notre DataFrame sélectionné a " + str(df.shape[1]) + " colonnes.\n"
          "Il y a " + str(tableau_val_manquantes_ren_colonnes.shape[0]) +
          " colonnes qui ont des valeurs manquantes.")
    
    # Retourner le DataFrame avec les informations sur les valeurs manquantes
    return tableau_val_manquantes_ren_colonnes
```

```python
def colonnes_multivalues_uniques_par_contrat(df, colonne_reference):
    colonnes_multivalues_uniques = []

    # Grouper le DataFrame par la colonne de référence et les colonnes uniques, puis compter les valeurs uniques
    for col in df.columns:
        if col != colonne_reference:
            unique_vals = df.groupby(colonne_reference)[col].nunique()
            if any(unique_vals > 1):
                colonnes_multivalues_uniques.append(col)

    return colonnes_multivalues_uniques

# Utilisation de la fonction avec le DataFrame BCM_DATA et la colonne de référence 'REFERENCECONTRAT_'
colonnes_multivalues_uniques = colonnes_multivalues_uniques_par_contrat(BCM_DATA, 'REFERENCECONTRAT_')

# Affichage des noms des colonnes identifiées
print("Colonnes avec plus d'une seule valeur identique pour un seul contrat unique :")
for colonne in colonnes_multivalues_uniques:
    print(colonne)
```

```python
def plot_distribution(data, title, xlabel, ylabel, color='blue', bins=20):
    plt.figure(figsize=(10, 6))
    n, bins, patches = plt.hist(data, color=color, bins=bins)

    for rect in patches:
        height = rect.get_height()
        plt.annotate('{}'.format(int(height)), xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
def compter_clients_chevauchement_contrats(df):
    # Identifier les clients ayant plus d'un contrat
    clients_multiples = df[df['nombre_total_descontrats'] > 1]['CLIENTCODE_'].unique()
    
    # Initialiser le compteur de clients avec chevauchement
    nb_clients_chevauchement = 0
    
    for client in clients_multiples:
        # Filtrer les contrats du client
        contrats_client = df[df['CLIENTCODE_'] == client].sort_values(by='DATEACCORD_CONTRAT')
        
        # Vérifier s'il y a chevauchement entre les contrats
        for i in range(len(contrats_client) - 1):
            date_fin_contrat_precedent = contrats_client.iloc[i]['DATEFIN_CONTRAT']
            date_debut_contrat_suivant = contrats_client.iloc[i + 1]['DATEACCORD_CONTRAT']
            if date_fin_contrat_precedent > date_debut_contrat_suivant:
                nb_clients_chevauchement += 1
                break
    
    return nb_clients_chevauchement

# Appel de la fonction avec votre dataframe df_unique_contrat
nb_clients_chevauchement = compter_clients_chevauchement_contrats(df_unique_contrat)

# Afficher le résultat
print(f"Nombre total de clients dont les contrats en meme periode : {nb_clients_chevauchement}")

```

```python
def marquer_chevauchement_contrats(df):
    # Identifier les clients ayant plus d'un contrat
    clients_multiples = df[df['nombre_total_descontrats'] > 1]['CLIENTCODE_'].unique()
    
    # Initialiser la colonne de chevauchement avec 0
    df['chevauchement_contrat'] = 0
    
    for client in clients_multiples:
        # Filtrer les contrats du client
        contrats_client = df[df['CLIENTCODE_'] == client].sort_values(by='DATEACCORD_CONTRAT')
        
        # Vérifier s'il y a chevauchement entre les contrats
        chevauchement = 0
        for i in range(len(contrats_client) - 1):
            date_fin_contrat_precedent = contrats_client.iloc[i]['DATEFIN_CONTRAT']
            date_debut_contrat_suivant = contrats_client.iloc[i + 1]['DATEACCORD_CONTRAT']
            if date_fin_contrat_precedent > date_debut_contrat_suivant:
                chevauchement = 1
                break
        
        # Mettre à jour la colonne de chevauchement pour le client
        df.loc[df['CLIENTCODE_'] == client, 'chevauchement_contrat'] = chevauchement
    
    return df

# Appliquer la fonction à votre dataframe
df_unique_contrat = marquer_chevauchement_contrats(df_unique_contrat)

# Afficher un échantillon pour vérifier les résultats
print(df_unique_contrat[['CLIENTCODE_', 'chevauchement_contrat']].head())
```

```python
def compter_classification_contrats_chevauchement(df):
    # Initialiser les compteurs pour les différentes classifications
    sains_chevauchement = 0
    douteux_chevauchement = 0
    en_souffrance_chevauchement = 0
    
    sains_sans_chevauchement = 0
    douteux_sans_chevauchement = 0
    en_souffrance_sans_chevauchement = 0
    
    # Parcourir chaque client unique
    for client in df['CLIENTCODE_'].unique():
        # Filtrer les contrats du client
        contrats_client = df[df['CLIENTCODE_'] == client].sort_values(by='DATEACCORD_CONTRAT')
        
        # Vérifier s'il y a chevauchement entre les contrats
        chevauchement = False
        for i in range(len(contrats_client) - 1):
            date_fin_contrat_precedent = contrats_client.iloc[i]['DATEFIN_CONTRAT']
            date_debut_contrat_suivant = contrats_client.iloc[i + 1]['DATEACCORD_CONTRAT']
            if date_fin_contrat_precedent > date_debut_contrat_suivant:
                chevauchement = True
                break
        
        # Compter les classifications pour ce client
        if chevauchement:
            if 'SAINE' in contrats_client['Classification_Target'].values:
                sains_chevauchement += 1
            elif 'DOUTEUSE' in contrats_client['Classification_Target'].values:
                douteux_chevauchement += 1
            elif 'EN SOUFFRANCE' in contrats_client['Classification_Target'].values:
                en_souffrance_chevauchement += 1
        else:
            if 'SAINE' in contrats_client['Classification_Target'].values:
                sains_sans_chevauchement += 1
            elif 'DOUTEUSE' in contrats_client['Classification_Target'].values:
                douteux_sans_chevauchement += 1
            elif 'EN SOUFFRANCE' in contrats_client['Classification_Target'].values:
                en_souffrance_sans_chevauchement += 1
    
    # Retourner les résultats
    return (sains_chevauchement, douteux_chevauchement, en_souffrance_chevauchement,
            sains_sans_chevauchement, douteux_sans_chevauchement, en_souffrance_sans_chevauchement)
```

```python
# Appel de la fonction 
(sains_chevauchement, douteux_chevauchement, en_souffrance_chevauchement,
 sains_sans_chevauchement, douteux_sans_chevauchement, en_souffrance_sans_chevauchement) = compter_classification_contrats_chevauchement(df_unique_contrat)

# Afficher les résultats
print("Parmi les clients dont les contrats se chevauchent :")
print(f"   - {sains_chevauchement} sont classifiés comme 'sains'")
print(f"   - {douteux_chevauchement} sont classifiés comme 'douteux'")
print(f"   - {en_souffrance_chevauchement} sont classifiés comme 'en souffrance'")

print("\nParmi les clients sans chevauchement de contrat :")
print(f"   - {sains_sans_chevauchement} sont classifiés comme 'sains'")
print(f"   - {douteux_sans_chevauchement} sont classifiés comme 'douteux'")
print(f"   - {en_souffrance_sans_chevauchement} sont classifiés comme 'en souffrance'")
```

```python
def remplir_valeurs_manquantes_par_classe_majoritaire(df, colonne_cible, colonne_a_remplir):

    # Calculer les valeurs majoritaires pour chaque classe de la colonne cible
    valeurs_majoritaires_par_classe = df.groupby(colonne_cible)[colonne_a_remplir].apply(lambda x: x.value_counts().idxmax())
    
    # Remplir les valeurs manquantes avec les valeurs majoritaires correspondantes
    df[colonne_a_remplir] = df.apply(lambda row: valeurs_majoritaires_par_classe[row[colonne_cible]] 
                                     if pd.isnull(row[colonne_a_remplir]) else row[colonne_a_remplir], axis=1)
    
    return df
```

```python
# Liste des colonnes avec des valeurs manquantes
colonnes_valeurs_manquantes = df_unique_contrat.columns[df_unique_contrat.isnull().any()].tolist()

# Appliquer la fonction remplir_valeurs_manquantes_par_classe_majoritaire pour chaque colonne avec des valeurs manquantes
for colonne in colonnes_valeurs_manquantes:
    df_unique_contrat = remplir_valeurs_manquantes_par_classe_majoritaire(df_unique_contrat, 'Classification_Target', colonne)
```

```python
from scipy.stats import chi2_contingency

# Sélection des variables catégorielles (type object)
variables_categorielles = df_unique_contrat.select_dtypes(include='object').columns

# Initialisation de la liste des variables significatives et non significatives
variables_significatives = []
variables_nonsignificatives = []

# Définition de alpha
alpha = 0.05

# Boucle sur les variables catégorielles
for variable in variables_categorielles:
    # Création de la table de contingence
    table_de_contingence = pd.crosstab(df_unique_contrat[variable], df_unique_contrat['Classification_Target'])
    
    # Application du test du chi2 à la table de contingence
    chi2_statistique, p_value, _, _ = chi2_contingency(table_de_contingence)
    
    # Calcul de l'intensité du lien de Cramer (test d'intensité)
    n = table_de_contingence.sum().sum()  # Nombre total d'observations
    k = min(table_de_contingence.shape) - 1  # Nombre de catégories de la variable la plus large
    cramers_v = np.sqrt(chi2_statistique / (n * k))
    
    # Affichage de la table de contingence
    print(f"Table de contingence pour la variable '{variable}':")
    print(table_de_contingence)
    print("-----------------------------------------------------------")
    
    # Vérification de la p-valeur
    if p_value < alpha:
        # La variable est significative, donc on l'ajoute à la liste des variables significatives
        variables_significatives.append((variable, cramers_v))
    else:
        # La variable est non significative, donc on l'ajoute à la liste des variables non significatives
        variables_nonsignificatives.append((variable, cramers_v))
```

```python
# Affichage des variables significatives
print("Variables significatives après le test chi2 (p-value < 0.05):")
for var, cramer_v in variables_significatives:
    print(f"Variable : {var}, Test d'intensité (Cramer's V) : {cramer_v}")
```

```python
import pandas as pd
from scipy.stats import kruskal
# Sélection des variables numériques
variables_numeriques = df_unique_contrat.select_dtypes(exclude='object').columns

# Variable catégorielle à trois modalités
variable_qualitative = 'Classification_Target'

# Initialisation des listes des variables significatives et non significatives
variables_significatives = []
variables_nonsignificatives = []

# Définition de alpha
alpha = 0.05

# Boucle sur les variables numériques
for variable_num in variables_numeriques:
    # Application du test de Kruskal-Wallis pour chaque variable numérique par rapport à la variable catégorielle
    groupe_1 = df_unique_contrat[df_unique_contrat[variable_qualitative] == df_unique_contrat[variable_qualitative].unique()[0]][variable_num]
    groupe_2 = df_unique_contrat[df_unique_contrat[variable_qualitative] == df_unique_contrat[variable_qualitative].unique()[1]][variable_num]
    groupe_3 = df_unique_contrat[df_unique_contrat[variable_qualitative] == df_unique_contrat[variable_qualitative].unique()[2]][variable_num]

    # Vérifier s'il y a au moins deux groupes distincts
    if len(groupe_1.unique()) < 2 or len(groupe_2.unique()) < 2 or len(groupe_3.unique()) < 2:
        print(f"Variable {variable_num} ne peut pas être testée avec Kruskal-Wallis : moins de deux groupes distincts.")
        continue
    
    # Application du test de Kruskal-Wallis
    kruskal_statistique, p_value = kruskal(groupe_1, groupe_2, groupe_3)
    
    # Affichage de la variable
    print(f"Variable numérique : {variable_num}")
    print(f"Statistique de Kruskal-Wallis : {kruskal_statistique}")
    print(f"p-value : {p_value}")
    print("-----------------------------------------------------------")
    
    # Vérification de la p-valeur
    if p_value < alpha:
        # La variable est significative, donc on l'ajoute à la liste des variables significatives
        variables_significatives.append((variable_num, kruskal_statistique))
    else:
        # La variable est non significative, donc on l'ajoute à la liste des variables non significatives
        variables_nonsignificatives.append((variable_num, kruskal_statistique))

# Affichage des variables significatives
print("Variables numériques significatives après le test de Kruskal-Wallis (p-value < 0.05):")
for var, kruskal_stat in variables_significatives:
    print(f"Variable : {var}, Statistique de Kruskal-Wallis : {kruskal_stat}")

# Liste des variables non significatives
colonnes_nonsignificatives = [var[0] for var in variables_nonsignificatives]
print("Colonnes numériques non significatives après le test de Kruskal-Wallis:")
print(colonnes_nonsignificatives)
```
